```python
import pandas as pd
import numpy as np

train = pd.read_csv("train.csv", index_col=0)
test = pd.read_csv("test.csv",index_col=0)
sample_submit = pd.read_csv("sample_submit.csv",index_col=0,header = None)
```

```python
data = pd.concat([train,test])
```

```python
data["age"] = data["age"].fillna(data["age"].mean())
data["embarked"] = data["embarked"].fillna(data["embarked"].mode())
```

```python
data = pd.get_dummies(data)
```

```python
dummy_cols = ['sex_female', 'sex_male', 'embarked_C', 'embarked_Q', 'embarked_S']
data[dummy_cols] = data[dummy_cols].astype(int)
```

```python
train = data.loc[train.index]
test = data.loc[test.index]
test = test.drop(["survived"], axis=1)
```

```python
train.head()
```

```python
y = train["survived"]
X = train.drop(["survived"], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X,y)
```

```python
pred = model.predict_proba(test)[:,1]
```

```python
print(pred[:5])
```

```python
sample_submit[1] = pred
sample_submit.to_csv('submit_2.csv', header=None)
```
