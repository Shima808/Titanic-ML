```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier


train = pd.read_csv("train.csv", index_col=0)
test = pd.read_csv("test.csv", index_col=0)
sample_submit = pd.read_csv("sample_submit.csv", index_col=0, header=None)
# データ準備（基本特徴量のみ）
data = pd.concat([train, test])
data["age"] = data["age"].fillna(data["age"].median())
data["embarked"] = data["embarked"].fillna(data["embarked"].mode()[0])
data["fare"] = data["fare"].fillna(data["fare"].median())
data = pd.get_dummies(data, columns=['sex', 'embarked'], dtype=int)

train_data = data.loc[train.index]
test_data = data.loc[test.index].drop(["survived"], axis=1)
y = train_data["survived"]
X = train_data.drop(["survived"], axis=1)

# 複数seedでGB訓練して平均
preds = []
for seed in [42, 0, 1, 7, 99, 123, 2024]:
    model = GradientBoostingClassifier(
        learning_rate=0.01, max_depth=4, min_samples_split=5,
        n_estimators=300, subsample=0.8, random_state=seed
    )
    model.fit(X, y)
    preds.append(model.predict_proba(test_data)[:, 1])

pred = np.mean(preds, axis=0)

sample_submit[1] = pred
sample_submit.to_csv('submit_8.csv', header=None)

from google.colab import files
files.download('submit_8.csv')
```
